```python
# 1. Import library
import pandas as pd
from pymongo import MongoClient
import os
from dotenv import load_dotenv

# 2. Load .env dan sambungkan ke MongoDB
load_dotenv()
client = MongoClient(os.getenv("MONGO_URI"))
db = client["cuaca_db"]
collection = db["data_cuaca"]

# 3. Ambil data dari MongoDB
data_list = list(collection.find())

# 4. Masukkan ke DataFrame
df = pd.DataFrame(data_list)
df.head()
```

```python
# Ubah kolom timestamp menjadi datetime
df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")

# Rename agar lebih rapi dan standar
df = df.rename(columns={
    "city": "kota",
    "temperature": "suhu",
    "humidity": "kelembaban",
    "weather": "cuaca"
})

# Ambil kolom penting
df = df[["timestamp", "kota", "suhu", "kelembaban", "cuaca"]]
df.head()
```
