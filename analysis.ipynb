```python
# 1. Import library
import pandas as pd
from pymongo import MongoClient
import os
from dotenv import load_dotenv
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load .env dan sambungkan ke MongoDB
load_dotenv()
client = MongoClient(os.getenv("MONGO_URI"))
db = client["cuaca_db"]
collection = db["data_cuaca"]

# 3. Ambil data dari MongoDB
data_list = list(collection.find())

# 4. Masukkan ke DataFrame
df = pd.DataFrame(data_list)
df.head()
```

```python
# Rename agar lebih rapi dan standar
df = df.rename(columns={
    "city": "kota",
    "temperature": "suhu",
    "humidity": "kelembaban",
    "pressure": "tekanan",
    "weather": "cuaca",
    "wind_speed": "kecepatan_angin",
    "local_time": "waktu_lokal"
})
# Ambil kolom penting
df = df[["waktu_lokal", "kota", "suhu", "kelembaban", "tekanan", "kecepatan_angin", "cuaca"]]
df.head()
```
